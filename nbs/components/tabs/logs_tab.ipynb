# Logs Tab

> Log tailing view with auto-scroll.

```python
#| default_exp components.tabs.logs_tab
```

```python
#| export
from fasthtml.common import Div, Pre, Code, Script, FT

from cjm_fasthtml_daisyui.utilities.semantic_colors import bg_dui, text_dui
from cjm_fasthtml_tailwind.utilities.spacing import p
from cjm_fasthtml_tailwind.utilities.sizing import max_h, w
from cjm_fasthtml_tailwind.utilities.typography import font_size, font_family
from cjm_fasthtml_tailwind.utilities.layout import overflow
from cjm_fasthtml_tailwind.utilities.borders import rounded
from cjm_fasthtml_tailwind.core.base import combine_classes

from cjm_fasthtml_job_monitor.html_ids import JobMonitorHtmlIds
```

## render_logs_tab

```python
#| export
def render_logs_tab(
    ids: JobMonitorHtmlIds,  # Element IDs
    logs: str = '',          # Log text content
) -> FT:  # Logs tab content
    """Render logs tab with auto-scroll to bottom."""
    return Div(
        Pre(
            Code(logs or 'No logs available.'),
            id=ids.log_container,
            cls=combine_classes(
                bg_dui.base_200, rounded.lg, p(3),
                max_h(64), overflow.y.auto, w.full,
                font_size.xs, font_family.mono, text_dui.base_content,
            ),
        ),
        # Auto-scroll to bottom on each update
        Script(f"""
        requestAnimationFrame(() => {{
            const el = document.getElementById('{ids.log_container}');
            if (el) el.scrollTop = el.scrollHeight;
        }});
        """),
        cls=p(4),
    )
```

```python
# Test render_logs_tab
from cjm_fasthtml_job_monitor.html_ids import JobMonitorHtmlIds

ids = JobMonitorHtmlIds(prefix="jm")

# With content
result = render_logs_tab(ids, logs="Line 1\nLine 2\nLine 3")
pre = result.children[0]
assert pre.attrs['id'] == 'jm-log-container'
assert 'max-h-64' in pre.attrs['class']
assert 'overflow-y-auto' in pre.attrs['class']
assert 'font-mono' in pre.attrs['class']
code = pre.children[0]
assert 'Line 1' in str(code)
print("render_logs_tab with content: OK")

# Empty
result_empty = render_logs_tab(ids, logs='')
code_empty = result_empty.children[0].children[0]
assert 'No logs available.' in str(code_empty)
print("render_logs_tab empty: OK")
```
