> Semi-transparent content overlay with loading spinner.

```python
#| export
from fasthtml.common import Div, Span, FT

from cjm_fasthtml_daisyui.components.feedback.loading import loading, loading_styles, loading_sizes
from cjm_fasthtml_daisyui.utilities.semantic_colors import bg_dui
from cjm_fasthtml_tailwind.utilities.layout import position, inset, z, display_tw
from cjm_fasthtml_tailwind.utilities.flexbox_and_grid import flex_display, items, justify
from cjm_fasthtml_tailwind.utilities.borders import rounded
from cjm_fasthtml_tailwind.utilities.effects import opacity
from cjm_fasthtml_tailwind.core.base import combine_classes

from cjm_fasthtml_job_monitor.html_ids import JobMonitorHtmlIds
from cjm_fasthtml_job_monitor.models import JobMonitorConfig
```

## render_job_overlay

Semi-transparent overlay that covers the consumer's content area during job execution.
The overlay target container must have `position: relative` for this to work.

```python
#| export
def render_job_overlay(
    ids: JobMonitorHtmlIds,       # Element IDs
    config: JobMonitorConfig,     # Display config (for z-index)
) -> FT:  # Overlay div with centered spinner
    """Render semi-transparent overlay with loading spinner."""
    return Div(
        Span(cls=combine_classes(loading, loading_styles.spinner, loading_sizes.lg)),
        id=ids.overlay,
        cls=combine_classes(
            position.absolute, inset(0), z(config.overlay_z_index),
            flex_display, items.center, justify.center,
            bg_dui.base_300, opacity(50),
            rounded.lg,
        ),
    )
```

```python
# Test render_job_overlay
from cjm_fasthtml_job_monitor.html_ids import JobMonitorHtmlIds
from cjm_fasthtml_job_monitor.models import JobMonitorConfig

ids = JobMonitorHtmlIds(prefix="jm")
config = JobMonitorConfig()

overlay = render_job_overlay(ids, config)
assert overlay.attrs['id'] == 'jm-overlay'
cls = overlay.attrs['class']
assert 'absolute' in cls
assert 'inset-0' in cls
assert 'z-10' in cls
assert 'opacity-50' in cls
assert 'bg-base-300' in cls
# Spinner child
spinner = overlay.children[0]
assert 'loading' in spinner.attrs['class']
assert 'loading-spinner' in spinner.attrs['class']
assert 'loading-lg' in spinner.attrs['class']
print("render_job_overlay: OK")
```

## render_job_overlay_placeholder

Empty placeholder that replaces the overlay on completion/cancel/fail. Swapping this in effectively removes the overlay.

The placeholder is `display:none` so it contributes no layout slot (no flex gap, no grid track) to the overlay target's layout tree. The active `render_job_overlay` is `position: absolute`, so neither state affects sibling element placement — the `*-overlay` container stays structurally invisible regardless of job state.

```python
#| export
def render_job_overlay_placeholder(
    ids: JobMonitorHtmlIds,  # Element IDs
) -> FT:  # Empty div with overlay ID (display:none)
    """Render empty placeholder (swapping this in removes the overlay).

    The placeholder has `display:none` so it contributes no layout
    (no flex gap slot, no grid track) to the overlay target's layout tree.
    The active `render_job_overlay` uses `position: absolute` for the same
    guarantee; both states keep the overlay container structurally invisible.
    """
    return Div(
        id=ids.overlay,
        cls=combine_classes(display_tw.hidden),
    )
```

```python
# Test render_job_overlay_placeholder
placeholder = render_job_overlay_placeholder(ids)
assert placeholder.attrs['id'] == 'jm-overlay'
assert len(placeholder.children) == 0
# Regression guard: placeholder must be display:none so it doesn't
# contribute a flex gap slot or grid track in the overlay target's layout.
assert 'hidden' in placeholder.attrs.get('class', '')
print("render_job_overlay_placeholder: OK")
```
